### Упражнение 3 *
Напишите функцию, считающую коэффициенты МНК для набора измерений x и y.

```python
def MNK():
    x = list(map(float, input('Введите значения x разделяя их пробелами\n').split()))
    y = list(map(float, input('Введите значения y разделяя их пробелами\n').split()))
    if len(x) == len(y):
        n = len(x)
        a = (n * sum(i * j for i, j in zip(x, y)) - sum(x) * sum(y)) / (n * sum(i ** 2 for i in x) - sum(x) ** 2)
        b = (sum(y) - a * sum(x)) / n
        return a, b
    else:
        return 'Ошибка в введённых данных'

print(MNK())
```

### Удобно решать через zip

Функция zip() в Python создает итератор, который объединяет элементы из нескольких источников данных. 
Эта функция работает со списками, кортежами, множествами и словарями для создания списков или кортежей, 
включающих все эти данные.
Эта функция предустоновленная!
Функция zip возвращает итератор кортежей, где i-й кортеж содержит i-й элемент каждого из списков.

```python
marks_numbers = [2, 3, 4, 5]
marks_names = ["Неуд", "Уд", "Хор", "Отл"]

zipped_values = zip(marks_numbers, marks_names)

print(zipped_values)
```

В отличие от Python 2, функция zip в Python 3 возвращает итератор.
Цель этого - сэкономить память, генерируя элементы итератора только по мере необходимости, 
а не помещая все это в память сразу. Если вы хотите по мере кода использовать свой заархивированный объект,
просто создайте из него список

zip-элемент конвертируется в список так, чтобы его можно было выводить и использовать как список.

```python
zipped_list = list(zipped_values)

print(zipped_list)
```

# Разница между глубоким и поверхностным копированием

Стандартные операторы присваивания Python, такие как = и := не справляются с этой задачей, 
а лишь только привязывают новые имена к уже имеющимся объектам в ОЗУ. По сути, использование 
стандартных операторов присваивания для вышеназванных объектов данных может иметь смысл лишь тогда, 
когда эти объекты являются неизменяемыми.

```python
def yell(text):
    return text.upper() + '!'
```

```python
bark = yell #принимается объект функции, на который ссылается yell, и создается второе имя, указывающее на него, bark
```

```python
bark('woof')
```

Объекты функций и их имена - это две отдельные проблемы. Можно удалить, поменять исходную функцию.
Поскольку другое имя (bark) по-прежнему указывает на базовую функцию, вы все равно можете вызвать функцию через него:

```python
def yell(text):
    return text.upper() + '?!'
```

```python
yell ('woof')
```

```python
bark('woof')
```

При глубоком копирование новый объект хранит копии значений объекта, 
тогда как при поверхностным копирование новый объект хранить ссылки на исходный адрес памяти
Глубокая копия не отражает изменения, внесенные в новый/скопированный объект в исходном объекте; 
в то время как поверхностная копия отражает

Модуль позволяет создавать слабые ссылки на объекты.

```python
import weakref

def yell(text):
    return text.upper() + '!'

bark = weakref.ref(yell)

del yell

if bark() is not None:
    print(bark()('woof'))  
else:
    print("Функция удалена")
```

Проблема заключается в реализации копирования для разных типов объектов

```python
import copy
```

Вложенные списки

```python
original_list = [[1, 2], [3, 4]]
copy_list = copy.copy(original_list)

original_list.append([5, 6])
original_list[0][0] = 'X'

print(original_list) 
print(copy_list) 
```

```python
original_list = [[1, 2], [3, 4]]
deepcopy_list = copy.deepcopy(original_list)

original_list.append([5, 6])
original_list[0][0] = 'X'

print(original_list)  
print(deepcopy_list)
```

Словарь

```python
original_dict = {'a': [1, 2], 'b': [3, 4]}
copy_dict = copy.copy(original_dict)
deepcopy_dict = copy.deepcopy(original_dict)

original_dict['a'].append(5)
original_dict['c'] = [6, 7]

print(original_dict) 
print(copy_dict)  
print(deepcopy_dict) 
```

```python
A = [10, 9, 10] 
B = copy.deepcopy(A) 

print(A)  
print(B)  
```

```python
B[0] = 3

print("Original List: ")  
print(A)  
print("Deep Copy:")  
print(B)  
```

```python
C = copy.copy(A)

print(A)  
print(C)  
```

```python
C[0] = 5

print("Original List: ")  
print(A)  
print("Shallow Copy:")  
print(C)  
```

A - список, в нем только целые числа, а не списки или другие изменяемые объекты.

Когда делаем C = copy.copy(A), создается поверхностная копия списка A. 
Это означает, что создается новый список C, содержащий ссылки на те же самые целые числа, что и список A.

НО!!!, целые числа являются неизменяемыми объектами, изменение значения в списке C не влияет на список A. 
Когда изменяем значения в списке C то меняем не сам целочисленный объект, а только создаем новую ссылку на 
уже другое целое число.

```python
def spiral_nm(n, m):
    dx, dy = 1, 0 # приращения
    x, y = 0, 0 # нач коорд
    arr = [[0] * m for _ in range(n)]
    for i in range(1, n*m+1):
        arr[x][y] = i 
        new_x, new_y = x+dx, y+dy
        if 0 <= new_x < n and 0 <= new_y < m and not arr[new_x][new_y]:  #проверка не залезли ли мы не туда))
            x, y = new_x, new_y
        else:
            dx, dy = -dy, dx # если что, меняем на противоположное
            x, y = x+dx, y+dy
    for i, col in enumerate(zip(*arr)):
        print(*[j * (i+1) for j in col])


n, m = 7, 6
spiral_nm(n, m)
```

```python
print('В какой стране вы живете?')
country = input()

if country in Capitals:
    print('Столица вашей страны', Capitals[country])
else:
    print('Как называется столица вашей страны?')
    city = input()
    Capitals[country] = city
```
