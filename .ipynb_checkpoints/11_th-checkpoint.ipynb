# Декораторы в классах

Декоратор класса — модифицирует поведение класса, добавляя или удаляя свойства, методы или атрибуты, модифицирует метод класса, изменяя его поведение или добавляя дополнительную функциональность.

Для начала вспомним что такое декоратор на примере функций

### Декораторы функций

```python
def my_decorator(function_to_decorate):
     # Внутри себя декоратор определяет функцию-"обёртку". Она будет обёрнута вокруг декорируемой,
     # получая возможность исполнять произвольный код до и после неё.
    def wrapper():
        print("Я - код, который отработает до вызова функции")
        function_to_decorate() # Сама функция
        print("А я - код, срабатывающий после")
     # Вернём эту функцию
    return wrapper
    
def alone_function():
     print("Наша функция для изменения")

# Однако, чтобы изменить её поведение, мы можем декорировать её, то есть просто передать декоратору,
# который обернет исходную функцию в любой код, который нам потребуется, и вернёт новую,
# готовую к использованию функцию:
function_decorated = my_decorator(alone_function)
function_decorated()
```

Когда мы вызываем имя и документацию декорированной функции, мы понимаем, что они были заменены значениями из функции декоратора. Это плохо, поскольку мы не можем перезаписывать все наши имена функций и документацию каждый раз при использовании какого-либо декоратора.

Мы уже говорили о кое-каких декораторах, но не о декораторах для создания еще большего количества декораторов. Одним из таких декораторов является functools.wraps.

```python
import functools

def uppercase(func):
    @functools.wraps(func)
    def wrapper():
        result = func()
        modified_result = result.upper()
        return modified_result
    return wrapper

@uppercase
def greet_():
    return 'Hello!'

print(greet_())
print(greet_.__name__)
print(greet_.__doc__)
```

```python
import functools

def uppercase(func):
    @functools.wraps(func)
    def wrapper():
        result = func()
        modified_result = result.upper()
        return modified_result
    return wrapper

@uppercase
def greet():
    return 'Hello!'

print(greet())
print(greet.__name__)
print(greet.__doc__)

greet = greet.__wrapped__

print(greet()) 
print(greet.__name__)
print(greet.__doc__)
```

Единственная задача функции wraps – это копирование имени, документации, списка аргументов и т.д., для предотвращения перезаписи. Если учесть, что wraps – это тоже декоратор, можно просто добавить его в наш код, и проблема решена! 

 код снимает декоратор с помощью атрибута __wrapped__, который возвращает ссылку на исходную функцию greet до применения декоратора.

```python
import functools

def uppercase(func):
    def wrapper():
        result = func()
        modified_result = result.upper()
        return modified_result
    return wrapper

@uppercase
def greet():
    return 'Hello!'

print(greet())
print(greet.__name__)
print(greet.__doc__)

greet = greet.__wrapped__

print(greet()) 
print(greet.__name__)
print(greet.__doc__)
```

Я пытаюсь использовать атрибут __wrapped__  с функцией greet, но этот атрибут существует только в том случае, если функция была декорирована с помощью functools.wraps.

```python
def uppercase(func):
    def wrapper():
        result = func()
        modified_result = result.upper()
        return modified_result
    return wrapper

def greet():
    return 'Hello!'

original_greet = greet  # сохраняем ссылку на исходную функцию

greet = uppercase(greet)

print(greet())
print(greet.__name__)
print(greet.__doc__)

print(original_greet()) 
print(original_greet.__name__)
print(original_greet.__doc__)
```

### Декораторы в классах

Создадим класс Cat, в котором объявим три атрибута: порода — breed, цвет — color и возраст — age. А ещё добавим 2 метода

```python
class Cat():
    def __init__(self, breed, color, age):
        self.breed = breed
        self.color = color
        self.age = age
    
    def info(self):
        print(f"Порода: {self.breed}, Цвет: {self.color}, Возраст: {self.age}")

    def meow(self):
        print('Мяу!')

cat1 = Cat('Британская', 'Серая', 4)

cat1.info()
cat1.meow()

cat1.age = 5
cat1.info()
```

```python
class Cat():
    def __init__(self, breed, color, age):
        self._breed = breed
        self._color = color
        self._age = age

    @property
    def breed(self):
        return self._breed

    @property
    def color(self):
        return self._color

    @property
    def age(self):
        return self._age

    def info(self):
        print(f"Порода: {self.breed}, Цвет: {self.color}, Возраст: {self.age}")

    def meow(self):
        print('Мяу!')

cat1 = Cat('Британская', 'Серая', 4)

cat1.info()
cat1.meow()
cat1.age = 5
cat1.info()
```

Код стал выглядеть немного сложнее. 

Сначала мы сделали все атрибуты закрытыми с помощью символа _. Он говорит интерпретатору, что эта переменная будет доступна только внутри методов класса.

Нам всё ещё нужно получать доступ к атрибутам, поэтому мы предоставляем его через @property и объявляем для каждого атрибута свой метод — breed, color, age. В каждом из этих методов мы возвращаем значение нашего закрытого атрибута. Это делает его доступным только для чтения.


### Getter and setter

```python
class Person:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
    @property
    def full_name(self):
        return self.name + ' ' + self.surname

olya = Person('Olga', 'Kolesnikova')
print(olya.full_name)
olya.name = 'Helen'
print(olya.full_name)
```

full_name вычисляется динамически. И если мы поменяем name, то full_name также изменится.
Такой метод, возвращающий динамически вычисляемое значение, называется геттером (getter).


```python
olya.full_name = 'Olga Kolesnikova'
```

```python
class Person:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname

    @property
    def full_name(self):
        return self.name + ' ' + self.surname

    @full_name.setter
    def full_name(self, other):
        self.name, self.surname = other.split(' ')

olya = Person('Olga', 'Kolesnikova')
print(olya.full_name)
olya.full_name = 'Palina Shishkina'
print(olya.full_name)
print(olya.name) 
print(olya.surname)
```

А пусть мы все-таки хотим что-то менять, но с ограничениями.

```python
class Cat():
    def __init__(self, breed, color, age):
        self._breed = breed
        self._color = color
        self._age = age

    @property
    def breed(self):
        return self._breed
    
    @property
    def color(self):
        return self._color
    
    @property
    def age(self):
        return self._age
  
    @age.setter
    def age(self, new_age):
        if new_age > self._age:
            self._age = new_age
            return self._age
            
    def info(self):
        print(f"Порода: {self.breed}, Цвет: {self.color}, Возраст: {self.age}")

            
cat1 = Cat('Британская', 'Серая', 4)

cat1.info()
cat1.age = 5
cat1.info()
cat1.age = 3
cat1.info()
```

### Обратно к теме

```python
class HomeCat(Cat):
    def __init__(self, breed, color, age, owner):
        super().__init__(breed, color, age)
        self._owner = owner
  
    @property
    def owner(self):
        return self._owner

cat2 = HomeCat('Сиамская', 'Белая', 3, 'Оля')

print(cat2.owner)
print(cat2.breed)
```

Кроме атрибутов для класса-родителя у класса-потомка есть собственный атрибут: «хозяин» — owner.

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __lt__(self, other):
        return self.value < other.value

    def __le__(self, other):
        return self.value <= other.value

    def __eq__(self, other):
        return self.value == other.value

    def __ne__(self, other):
        return self.value!= other.value

    def __gt__(self, other):
        return self.value > other.value

    def __ge__(self, other):
        return self.value >= other.value
```

Имена self и other ни что другое, как договоренность между програмистами — вы можете использовать другие названия, какие вам угодны. 
(Хотя я вам это не рекомендую, когда хотите, чтобы другие программисты не имели трудности с пониманием вашего кода.)

self значит я, в смысле «я, текущий объект».
other значит другой, в смысле «другой операнд (после применения метода к первому)».

less-than и т.д.

```python
class Number:
    def __init__(self, value):
        self.value = value

a = Number(5)
b = Number(3)

print(a + b)
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value

a = Number(5)
b = Number(3)

print(a + b)
```

Декоратор @total_ordering из модуля functools используется для генерации недостающих методов сравнения для класса Python на основе тех, которые определены.

```python
from functools import total_ordering


@total_ordering
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __eq__(self, other):
        return self.grade == other.grade

    def __lt__(self, other):
        return self.grade < other.grade


student1 = Student("Alice", 85)
student2 = Student("Bob", 75)
student3 = Student("Charlie", 85)

print(student1 < student2)  
print(student1 > student2)  
print(student1 == student3)  
print(student1 <= student3)
print(student3 >= student2) 
```

### Ваши декораторы классов

Декоратор класса: Декоратор класса, который модифицирует поведение класса, добавляя или удаляя свойства, методы или атрибуты.

Декоратор метода класса: Декоратор Python, который модифицирует метод класса, изменяя его поведение или добавляя дополнительную функциональность.

Декоратор свойства класса: Декоратор класса, который модифицирует свойство класса, добавляя или удаляя атрибуты или методы.

```python
def add_run_method(cls):
    def run(self):
        print(f"{self.name} бежит.")
    cls.run = run
    return cls

@add_run_method
class DomesticAnimal:
    def __init__(self, name, age, species):
        self.name = name
        self.age = age
        self.species = species
    def info(self):
        print(f"Имя: {self.name}, Возраст: {self.age}, Вид: {self.species}")

dog = DomesticAnimal("Шарик", 3, "собака")

dog.run()  
```

```python
def add_info_to_eat_method(func):
    def wrapper(self):
        print(f"Метод eat вызван для {self.name}.")
        func(self)
    return wrapper

class DomesticAnimal:
    def __init__(self, name, age, species):
        self.name = name
        self.age = age
        self.species = species

    @add_info_to_eat_method
    def eat(self):
        print(f"{self.name} кушает.")
        
dog = DomesticAnimal("Шарик", 3, "собака")
dog.eat()  
```

```python
def add_weight_property(cls):
    def get_weight(self):
        return self._weight
    def set_weight(self, value):
        self._weight = value
    cls.weight = property(get_weight, set_weight)
    return cls
  

@add_weight_property
class DomesticAnimal:
    def __init__(self, name, age, species):
        self.name = name
        self.age = age
        self.species = species

dog = DomesticAnimal("Шарик", 3, "собака")
dog.weight = 10
print(dog.weight) 
```
