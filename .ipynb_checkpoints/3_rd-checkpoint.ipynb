#### Упражнение. Групповое обращение *

Рассмотрим операцию группового обращения строки. Строка разбивается на N групп символов равной длины, после чего в каждой группе порядок символов меняется на обратный. Зная количество групп в строке, необходимо найти ее обращение.

Вход. Количество групп G в строке и сама строка. Длина строки кратна G. Строка может содержать только буквы и цифры.

Выход. Вывести групповое обращение строки.

|Пример входа|Пример выхода|
|-|-|
|3 ABCEHSHSH|CBASHEHSH|
|6 AB0TAEL0ANANAB0EVO0SANAN|EAT0BANANA0LOVE0BANANAS0|

```python
g, s = input().split()
g = int(g)
s = list(s)

result = []
for i in range(0, len(s), g):
    result += s[i:i+g][::-1]

print(''.join(result))
```

### Вложенные функции

```python
def factorial(number):
    if not isinstance(number, int):
        raise TypeError("Число должно быть целым.")
    if number < 0:
        raise ValueError("Число должно быть неотрицательным.")
    def inner_factorial(number):
        if number <= 1:
            return 1
        return number * inner_factorial(number - 1)
    return inner_factorial(number)
```

```python
factorial(-3)
```

```python
def talk():
    age = int(input("Введите ваш возраст: "))
    def hello():
        return 'Привет'
    def hello_polite():
        return 'Здравствуйте'
    if age <= 30:
        return hello()
    else:
        return hello_polite()

talk()
```

Параметры функций в Python — это переменные, которым присваиваются значения в момент вызова функции.

### Именованные и позиционные параметры

Символ * позволяет установить, какие параметры будут именнованными - то есть такие параметры, которым можно передать значения только по имени. 
Все параметры, которые располагаются справа от символа *, получают значения только по имени:

```python
def print_person(name, *,  age, company):
    print(f"Name: {name}  Age: {age}  Company: {company}")
 
 
print_person("Bob", age = 18, company ="Microsoft")  
```

Если наоборот надо определить параметры, которым можно передавать значения только по позиции, то есть позиционные параметры, то можно использовать символ /: все параметры, которые идут до символа / , являются позиционными и могут получать значения только по позиции

```python
def print_person(name, /, age, company="Microsoft"):
    print(f"Name: {name}  Age: {age}  Company: {company}")
 
 
print_person("Tom", company="JetBrains", age = 24)
```

##### Ось (axis) 0. Это первая ось, которая соответствует направлению вниз по строкам. Если у нас двумерный набор данных и мы просуммируем элементы вдоль оси 0, то получим сумму элементов каждого столбца.
##### Ось (axis) 1. Это вторая ось, которая проходит вдоль второго измерения массива. В двумерной структуре она соответствует направлению по столбцам. Если мы просуммируем элементы вдоль оси 1, то в результате получим сумму элементов каждой строки.
##### Ось (axis) 2 и далее. Это дополнительные оси для массивов с тремя и более измерениями.

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

total_sum = arr.sum()
print(total_sum)

sum_along_axis0 = arr.sum(axis=0)
print(sum_along_axis0)

sum_along_axis1 = arr.sum(axis=1)
print(sum_along_axis1)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
       
mean_all = arr.mean()
print(mean_all)
               
# Среднее арифметическое элементов вдоль оси 0
mean_axis0 = arr.mean(axis=0)
print(mean_axis0)

# Среднее арифметическое элементов вдоль оси 1
mean_axis1 = arr.mean(axis=1)
print(mean_axis1)
```

Написать функцию, считающую N-ое число Фибоначчи (N вводится с клавиатуры).

```python
def fibonacci2(n):
    '''O(n)'''
    base1, base2 = 0, 1
    for i in range(n):
        base1, base2 = base2, base1 + base2
    return base1
```

```python
def print_triangle(size, symb):
    for i in range(1, size + 1):
        print(symb * i)
    for i in range(size - 1, 0, -1):
        print(symb * i)

print_triangle(5, '*')
```
